```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("data/default of credit card clients.xls")
```

```python
data.head(2)
```

```python
data.columns = data.iloc[0]
```

```python
data = data[1:]
```

```python
# counting the number of null in each coloums
data.isna().sum()
data[data.isna() == True].count()
```

```python
# counting the number of non null in each coloums
data[data.isna() == False].count()
```

```python
# shape of the data
data.shape
```

```python
# Count of Sex Destribution in the Data
fig = plt.figure(figsize=(12,6))

ax = data['SEX'].value_counts().plot.bar()
ax.set_xlabel('Number of Males and Females',fontsize=16)
ax.set_ylabel('Count',fontsize=16)
ax.set_title("Sex Destribution of the Data", fontsize=20)
```
